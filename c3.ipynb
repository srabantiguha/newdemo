```python
import pandas as pd
import re
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
stop_words = set(stopwords.words('english'))
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()
```

```python
df  = pd.read_csv(r'demo_text_data.csv')
df
```

```python
df.iloc[0,1]
```

```python
text = df.iloc[0,1]
re.sub(r'[^a-zA-Z\s]', '', str(text))
```

```python
df["text"] = df["text"].str.replace(r'[^a-zA-Z\s]', '', regex=True)
df
```

```python
df["clean_text"] = df["text"].str.replace(r'[^a-zA-Z\s]', '', regex=True)
df["clean_text"] = df["clean_text"].str.lower()
df
```

```python
df["tokens"] = df["clean_text"].apply(word_tokenize)
df
```

```python
stop_words = set(stopwords.words('english'))

df["no_stopwords"] = df["tokens"].apply(
    lambda words: [w for w in words if w not in stop_words]
)

df
```

```python
stop_words
```

```python
stemmer = PorterStemmer()

df["stemmed"] = df["no_stopwords"].apply(
    lambda words: [stemmer.stem(w) for w in words]
)
df
```

```python
lemmatizer = WordNetLemmatizer()

df["lemmatized"] = df["no_stopwords"].apply(
    lambda words: [lemmatizer.lemmatize(w) for w in words]
)
df
```
